```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt





df= pd.read_csv(r"\Users\USER\Desktop\Project\movies.csv")

```

```python
df.head(5)
```

```python
#check for missing data
for col in df.columns:
   percent_missing= np.mean(df[col].isnull())
   print(col,percent_missing,"%")
```

```python
#check for data type for the columns
df.dtypes
```

```python
#convert the data types of budget and gross to integer
df["budget"].replace(np.nan,0,inplace = True)
df["gross"].replace(np.nan,0,inplace = True)

df["budget"] = df["budget"].astype("int64")

df["gross"] = df["gross"].astype("int64")
```

```python
df.dtypes
```

```python
df.head(5)
```

```python
#arrang the data by the size of gross files  
df.sort_values(by=["gross"], inplace = False, ascending = False)
```

```python
#pd.set_option('display.max_rows',None)
```

```python
#remove duplicate rows
df.drop_duplicates()
```

```python
#scatter plot with budget to gross
plt.scatter(x =df["budget"],y=df["gross"])
plt.show
plt.title("Budget vs Gross earning")
plt.xlabel("Budget")
plt.ylabel("Gross earning")
```

```python
#regression plots with seaborn
sns.regplot(x = "budget",y="gross",data = df,line_kws={"color":"red"}) #scatter_kws
```

```python
#find the correlation
df.corr(numeric_only=True)
```

```python
#other techniques for correlation
df.corr(method = "spearman",numeric_only=True)
```

```python
#high correlation between budget and gross
```

```python
corr_matrix=df.corr(method = "pearson",numeric_only=True)
```

```python
#heat map for corelations
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Chart")
```

```python
#to find the corelation of roles with object data type, we convert them to numeric data type
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == "object"):
        df_numerized[col_name]=df_numerized[col_name].astype("category")
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
    
```

```python
#corr matrix for all
corr_matrix=df_numerized.corr(method = "pearson",numeric_only=True)
sns.heatmap(corr_matrix, annot=True, annot_kws={"size":5})

plt.show()
```

```python
#stack all this
stack=df_numerized.corr(method = "pearson",numeric_only=True).unstack()
stack
```
